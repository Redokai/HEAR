```python
import os
import pandas as pd
```

```python
output_folder = '..\Output\\'
```

```python
vid_df = pd.DataFrame(columns=['video','split','path','violence'])
videos =  os.listdir(output_folder)
for video in videos:
    video_folder = output_folder + video + '\\'
    splits = os.listdir(video_folder)
    for split in splits:
        violence = 'NEG' if 'NEG' in split else '?'
        split_path = video_folder + split + '\\'
        split_dic = {
            'video' : video,
            'split' : split,
            'path' : split_path,
            'violence' : violence
        }
        vid_df = vid_df.append(split_dic, ignore_index=True)
vid_df.head()
```

```python
optical_flow_folder = 'optical_flow\\'
frames_df = pd.DataFrame(columns=['video','split','frame','path','violence'])
for i, row in split_df.iterrows():
    video = row['video']
    split = row['split']
    folder = row['path']
    violence = row['violence']
    frame_files =  [f for f in os.listdir(folder) if f.split('.')[-1] == 'jpg']
    for frame in frame_files:
        frame_path = folder + frame
        d = {
            'video': video,
            'split': split,
            'optical flow interval': frame,
            'path': frame_path,
            'violence': violence
        }
        frames_df = frames_df.append(d, ignore_index=True)
        
    
```
