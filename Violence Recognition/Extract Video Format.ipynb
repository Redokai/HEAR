```python
import os
import pandas as pd
import cv2 as cv
import numpy as np
```

```python
file = 'manual_classification.csv'
accpected_frame_format = ['jpg']
```

```python
df = pd.read_csv(file, sep=';', index_col=0)
df['size'] = df['size'].astype(str)
df.head()
```

```python
for i, l in df.iterrows():
    files = [l.path + f for f in os.listdir(l.path) if f.split('.')[-1] in accpected_frame_format]
    if len(files) > 0:
        file_path = files[0]
        img = cv.imread(file_path)
        df.at[i,'size'] = str(img.shape[0]) + ' x ' + str(img.shape[1])
df.head()
```

```python
df['size']
```

```python
df = df[df['size'] != 'nan']
df['height'] = [f.split(' x ')[0] for f in list(df['size'])]
df['width'] = [f.split(' x ')[1] for f in list(df['size'])]
heights = np.array(list(df['height'].astype(int)))
widths = np.array(list(df['width'].astype(int)))
print('Min height :' + str(heights.min()))
print('Min width :' + str(widths.min()))
```
